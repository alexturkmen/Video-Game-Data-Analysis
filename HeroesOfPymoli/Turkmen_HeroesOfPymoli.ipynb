```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
csv_path = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
file_df = pd.read_csv(csv_path)
```

## Player Count

* Display the total number of players


```python
# Total Number of Players (some players buy more than one item therefore I have to use value_counts)

formatted_file_df = file_df.drop_duplicates(subset ="SN") 
total_number_of_players = formatted_file_df['SN'].count()

total_players = {'Total Number of Players':[total_number_of_players]}
total_players_df = pd.DataFrame(total_players)
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Total Number of Unique Items

number_of_unique_items = file_df["Item Name"].value_counts()
number_of_unique_items_df = pd.DataFrame(number_of_unique_items)
number_of_unique_items = number_of_unique_items_df["Item Name"].count()

# Average Purchase Price

average_purchase_price = round(file_df["Price"].mean(), 2)

formatted_average_purchase_price = '${0:,.2f}'.format(average_purchase_price)

# Number of Purchases

total_number_of_purchases = file_df["Price"].count()

# Total Revenue

total_revenue = file_df["Price"].sum()

formatted_total_revenue = '${0:,.2f}'.format(total_revenue)
```

```python
# Purchasing Analysis (Total)

purchasing_data = {'Number of Unique Items':[number_of_unique_items], 
                   'Average Price':[formatted_average_purchase_price], 
                   'Number of Purchases':[total_number_of_purchases], 
                   'Total Revenue': [formatted_total_revenue]
                  }

purchasing_analysis_total_df = pd.DataFrame(purchasing_data)
purchasing_analysis_total_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Count and Percentage of Male Players

male_players_df = formatted_file_df.loc[formatted_file_df["Gender"] == "Male"]
male_players_count = male_players_df["Gender"].count()

male_players_percentage = (male_players_count / formatted_file_df["Gender"].count()) * 100
formatted_male_players_percentage = '{0:,.2f}%'.format(male_players_percentage)


# Count and Percentage of Female Players

female_players_df = formatted_file_df.loc[formatted_file_df["Gender"] == "Female"]
female_players_count = female_players_df["Gender"].count()

female_players_percentage = (female_players_count / formatted_file_df["Gender"].count()) * 100
formatted_female_players_percentage = '{0:,.2f}%'.format(female_players_percentage)


# Count and Percentage of Other / Non-Disclosed Players

other_players_df = formatted_file_df.loc[formatted_file_df["Gender"] == "Other / Non-Disclosed"]
other_players_count = other_players_df["Gender"].count()

other_players_percentage = (other_players_count / formatted_file_df["Gender"].count()) * 100
formatted_other_players_percentage = '{0:,.2f}%'.format(other_players_percentage)


# Gender Demographics

gender_demographics = {'Total Count':[male_players_count, female_players_count, other_players_count], 
                       'Percentage of Players': [formatted_male_players_percentage, formatted_female_players_percentage, formatted_other_players_percentage]
                      }


gender_demographics_df = pd.DataFrame(gender_demographics)

gender_demographics_df_new = gender_demographics_df.rename(index={0: 'Male', 1: "Female", 2:"Other / Non-Disclosed"})
gender_demographics_df_new

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Use groupby based on genders

gender_group = file_df.groupby(["Gender"])
new_gender_group = gender_group.count()

# Delete unnecessary rows
del new_gender_group["SN"]
del new_gender_group["Age"]
del new_gender_group["Item ID"]
del new_gender_group["Item Name"]
del new_gender_group["Price"]

male_purchase_count_df = file_df.loc[file_df["Gender"] == "Male"]
female_purchase_count_df = file_df.loc[file_df["Gender"] == "Female"]
other_purchase_count_df = file_df.loc[file_df["Gender"] == "Other / Non-Disclosed"]

male_purchase_count = file_df.loc[file_df["Gender"] == "Male"]["Gender"].count()
female_purchase_count = file_df.loc[file_df["Gender"] == "Female"]["Gender"].count()
other_purchase_count = file_df.loc[file_df["Gender"] == "Other / Non-Disclosed"]["Gender"].count()

# Average Purchase Price

male_average_purchase_price = male_purchase_count_df['Price'].sum() / male_purchase_count

female_average_purchase_price = female_purchase_count_df['Price'].sum() / female_purchase_count

other_average_purchase_price = other_purchase_count_df['Price'].sum() / other_purchase_count


# Total Purchase Value

male_total_purchase_value = male_purchase_count_df['Price'].sum()

female_total_purchase_value = female_purchase_count_df['Price'].sum()

other_total_purchase_value = other_purchase_count_df['Price'].sum()


# Avg Total Purchase per Person

avg_total_purchase_per_male = male_total_purchase_value / male_players_count

avg_total_purchase_per_female = female_total_purchase_value / female_players_count

avg_total_purchase_per_other = other_total_purchase_value / other_players_count


# Adding new columns to the dataframe

av_pu_price = [female_average_purchase_price, male_average_purchase_price, other_average_purchase_price]
to_pu_value = [female_total_purchase_value, male_total_purchase_value, other_total_purchase_value]
av_to_pu_per_person = [avg_total_purchase_per_female, avg_total_purchase_per_male, avg_total_purchase_per_other]

new_gender_group.loc[:, "Average Purchase Price"] = av_pu_price
new_gender_group.loc[:, "Total Purchase Value"] = to_pu_value
new_gender_group.loc[:, "Avg Total Purchase per Person"] = av_to_pu_per_person


# Formatting
new_gender_group["Average Purchase Price"] = new_gender_group["Average Purchase Price"].astype(float).map("${:.2f}".format)
new_gender_group["Total Purchase Value"] = new_gender_group["Total Purchase Value"].astype(float).map("${:,.2f}".format)
new_gender_group["Avg Total Purchase per Person"] = new_gender_group["Avg Total Purchase per Person"].astype(float).map("${:.2f}".format)

final_gender_group = new_gender_group.rename(columns= {"Purchase ID": "Purchase Count"})

final_gender_group
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping the main dataframe(file_df) by "SN" column, and converting it into a new dataframe

SN_group = file_df.groupby(["SN"])
groupby_df = SN_group.sum()

# Deleting unnecessary columns

del groupby_df["Purchase ID"]
del groupby_df["Item ID"]
del groupby_df["Age"]

# Calculating Required Values

purc_id_counts = SN_group["Purchase ID"].count().values

groupby_df.loc[:, "Purchase Count"] = purc_id_counts

groupby_df.loc[:, "Average Purchase Price"] = round(groupby_df["Price"] / groupby_df["Purchase Count"], 2)

# Sorting the dataframe

top_spenders = groupby_df.sort_values(["Price"], ascending=False)

# Renaming Price column

top_spenders = top_spenders.rename(columns={"Price": "Total Purchase Value"})

# Organizing the dataframe

top_spenders_organized = top_spenders[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]

top_spenders_organized.head(5)

# Formatting the values in the columns

top_spenders_organized["Average Purchase Price"] = top_spenders_organized["Average Purchase Price"].astype(float).map("${:.2f}".format)
top_spenders_organized["Total Purchase Value"] = top_spenders_organized["Total Purchase Value"].astype(float).map("${:.2f}".format)

top_spenders_organized.head()
    
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping the main dataframe(file_df) by "Item ID" and "Item Name" columns, and converting it into a new dataframe

most_popular = file_df.groupby(["Item ID", "Item Name"]).count()

# Deleting unnecessary columns

del most_popular["Purchase ID"]
del most_popular["SN"]
del most_popular["Age"]
del most_popular["Gender"]

# Calculating Required Values and organizing the columns

mp_item_count = file_df.groupby(["Item ID"])["Purchase ID"].count().values
mp_item_sum = file_df.groupby(["Item ID"])["Price"].sum().values

most_popular_renamed = most_popular.rename(columns={"Price": "Purchase Count"})

most_popular_renamed.loc[:, "Total Purchase Value"] = mp_item_sum

most_popular_renamed.loc[:, "Average Purchase Price"] = round(most_popular_renamed["Total Purchase Value"] / most_popular_renamed["Purchase Count"], 2)

most_popular_renamed = most_popular_renamed.rename(columns={"Average Purchase Price": "Item Price"})

most_popular_items = most_popular_renamed[["Purchase Count", "Item Price", "Total Purchase Value"]]

# Sorting the Dataframe

formatted_most_popular_items = most_popular_items.sort_values(["Purchase Count"], ascending=False)


# Formatting the values in the columns

formatted_most_popular_items["Item Price"] = formatted_most_popular_items["Item Price"].astype(float).map("${:.2f}".format)
formatted_most_popular_items["Total Purchase Value"] = formatted_most_popular_items["Total Purchase Value"].astype(float).map("${:.2f}".format)

formatted_most_popular_items.head()
```
